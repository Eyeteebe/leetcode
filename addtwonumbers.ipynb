# AddTwoNumbers
- 给定两个非空链表来表示两个非负整数。位数按照逆序方式存储，它们的每个节点只存储单个数字。将两数相加返回一个新的链表。

- 你可以假设除了数字 0 之外，这两个数字都不会以零开头。

- 示例：
输入：(2 -> 4 -> 3) + (5 -> 6 -> 4),
输出：7 -> 0 -> 8,
原因：342 + 465 = 807

---

- 除了返回result[0]以外没有其它需要注意的，理清加法的几种不同情况即可。

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        carryDigit = 0
        result = []
        digits = 0
        while l1!=None and l2!=None:
            sum = l1.val + l2.val + carryDigit
            if sum>9:
                sum -= 10
                carryDigit = 1
            else:
                carryDigit = 0
            
            result.append(ListNode(sum))
            if digits > 0:
                result[-2].next = result[-1]
            
            digits += 1
            l1 = l1.next
            l2 = l2.next
            
        while l1!=None:
            sum = l1.val + carryDigit
            if sum>9:
                sum -= 10
                carryDigit = 1
            else:
                carryDigit = 0
                
            result.append(ListNode(sum))
            if digits > 0:
                result[-2].next = result[-1]
            
            digits += 1
            l1 = l1.next
                
        while l2!=None:
            sum = l2.val + carryDigit
            if sum>9:
                sum -= 10
                carryDigit = 1
            else:
                carryDigit = 0
                
            result.append(ListNode(sum))
            if digits > 0:
                result[-2].next = result[-1]
            
            digits += 1
            l2 = l2.next
                
        if 1==carryDigit:
            result.append(ListNode(carryDigit))
            carryDigit = 0
            if digits > 0:
                result[-2].next = result[-1]
        
        return result[0]
```
